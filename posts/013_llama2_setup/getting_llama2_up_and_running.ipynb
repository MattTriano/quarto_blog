# LLaMa 2

On July 19th, Facebook "open sourced" a collection of pretrained LLMs. 

## Setup

1. Request a download link for the weights by entering your email in [this form](https://ai.meta.com/resources/models-and-libraries/llama-downloads/). Meta will immediately email you a download link.

2. Clone the [`llama` repo](https://github.com/MattTriano/llama/tree/main)

3. Run the `download.sh` script and paste in your download link when prompted.

4. Create and activate new conda env

```bash
conda create --name llama2_env "python=3.10" jupyterlab
...
conda activate llama2_env
```

5. Get the command for installing the conda PyTorch package into your system from [here](https://pytorch.org/)

```bash
conda install  pytorch torchvision torchaudio pytorch-cuda=11.7 -c pytorch -c nvidia --override-channels
```

Don't forget to add on the `--override-channels` bit. If you've set up your base conda installation properly (ie you followed my [miniconda config instructions](../000_setting_up_miniconda/Setting_up_Miniconda)), conda will strictly prioritize installing from the conda-forge channel, but this install needs packages that are only available on the pytorch and nvidia conda channels.

6. (In the clones `/llama/` dir) Install the repo package via `python -m pip install -e .`
